# Google NQ Dataset

reading the original corpus dataset from google nq, filtering by the first 2000 documents and dropping the metadata column ../data_compressed/corpus_filtered.jsonl

```python
import pandas as pd
```

```python
df_nq = pd.read_json('../data/nq/corpus.jsonl', lines=True)
```

```python
df_nq.head()
```

```python
df_nq["title"].nunique()
```

```python
df_nq_filtered = df_nq[df_nq["title"].isin(df_nq["title"].unique()[:2000])]
```

```python
df_nq_filtered
```

```python
df_nq_filtered['title'].nunique()
```

```python
df_nq_filtered = df_nq_filtered.drop(columns=["metadata"])  
```

```python
df_nq_filtered.to_json('../data_compressed/corpus_filtered.jsonl', lines=True, orient='records')
```
