in the Query Tool clicking the database "vectordb"


````python
from sqlmodel import create_engine
from sqlalchemy import inspect

# 1. Define the connection string
# We use the 'psycopg' driver (v3) as requested in your previous snippet.
# Format: postgresql+driver://user:password@host:port/database_name
database_url = "postgresql+psycopg://bulacio@localhost:5432/vectordb"

# 2. Create the engine
# The engine is the central point of communication with the database
engine = create_engine(database_url)

# 3. Create an Inspector
# The inspector allows us to view database metadata (tables, columns, constraints)
inspector = inspect(engine)

# 4. Fetch and print table names
try:
    # get_table_names() retrieves a list of strings with the table names
    table_names = inspector.get_table_names()

    print(f"--- Tables found in database: {engine.url.database} ---")
    
    if table_names:
        for table in table_names:
            print(f"- {table}")
    else:
        print("No tables found in this database.")

except Exception as e:
    print(f"An error occurred while connecting: {e}")
````
